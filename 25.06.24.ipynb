25.06.25 - Function

```
def my_function():
  print("Hello From a Function")

my_function()

def my_function1(fname):
  print(fname + "Tefsnes")

my_function1("Emil")
my_function1("Toobias")
my_function1("Linus")

def my_function2(fname, Iname):
  print(fname + " " + Iname)

my_function2("Emil", "Refsnes")

#전달 인자의 갯수를 모를 경우 즉, n개의 인자를 받고싶을 때 사용 class는 tuple
def my_function3(*kids):
  print("The youngest child is" + kids[2])
  print("The youngest child is" + kids[1])
  print("The youngest child is" + kids[0])
  print(type(kids))

my_function3("Emil", "Tobias", "Linus")

#키워드 인자
def my_function4(child3, child2, child1):
  print("The Olderest child is" + child3)

my_function4(child1 = "Emil", child2 = "Tobias", child3 = "Linus")

#keyword arguments 즉, dict의 keyword를 받는다는 것.
def my_function5(**kid):
  print("His last name is" + kid["lname"])
  print(type(kid))

my_function5(fname = "Tobias", lname = "Refsnes")

def my_function6(country = "Norwat"):
  print("I am 신뢰에요" + country)

my_function6("sweden")
my_function6("india")
my_function6()
my_function6("Brazil")

def my_function7(food):
  for x in food:
    print(x)

fruits = ["apple", "banana", "cherry"]
my_function7(fruits)

def my_function8(x):
  return 7 * x

print(my_function8(3))
print(my_function8(4))
print(my_function8(5))

def myfunction9():
  pass

def my_function10(x, /):
  print(x)
my_function10(3)

def my_function11(*, x):
  print(x)
my_function11(x = 5)

# *가 붙는 순간 c = 7이라 명시를 해줘야 함
def my_function12(a, b, /, *, c,d):
  print(a+b+c+d)

my_function12(5,6, c = 7, d = 8)

#재귀함수
def tri_recursion(k):
  if(k > 0):
    result = k + tri_recursion(k - 1) #이걸로 계속 k값을 리턴해서 0꺄지함
    print(result)
  else:
    result = 0
  return result

print("Recursion Example Results: ")
tri_recursion(6)
```

Class

```
class MyClass:
  x = 5

p1 = MyClass()
print(p1.x)

class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("John", 36)
print(p1.name)
print(p1.age)

#__str__ = 문자열 객체 반환
class Person1:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def __str__(self):
    return f"{self.name}({self.age})"

p1 = Person1("John", 36)
print(p1)

#Obj
class Person2:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def myfunc(self):
    print("Hello my name is " + self.name)

p2 = Person2("John", 36)
p2.myfunc()

#self - Python의 첫 번째 매개변수, 굳이 self라 칭할 필요는 없으나 첫 번째 인자로 반드시 와야함
class Person3:
  def __init__(mysillyobj, name, age):
    mysillyobj.name = name
    mysillyobj.age = age

  def myfunc(abc):
    print("Hello my name is " + abc.name)

p3 = Person3("JohnG", 35)
p3.myfunc()
```
