```
import time
from ultralytics import YOLO
from google.colab import files
import glob
import os
import cv2

# ëª¨ë¸ íŒŒì¼ ì„ íƒ (ì˜ˆ: 'yolov11n.pt', 'yolov12n.pt')
MODEL_PATH = 'yolo11n.pt'  # ë˜ëŠ” 'yolov11n.pt'
model = YOLO(MODEL_PATH)
model.info()

# ë™ì˜ìƒ íŒŒì¼ ì—…ë¡œë“œ
uploaded = files.upload()
video_path = list(uploaded.keys())[0]

# FPS ì¸¡ì •ìš© ì¶”ë¡  ì‹¤í–‰
start = time.time()
results = model.predict(video_path, save=True)
end = time.time()

# í”„ë ˆì„ ìˆ˜ ë° ì†Œìš” ì‹œê°„
cap = cv2.VideoCapture(video_path)
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
cap.release()
total_time = end - start
fps = frame_count / total_time if total_time > 0 else 0
print(f"{MODEL_PATH} / í”„ë ˆì„ ìˆ˜: {frame_count} / ì´ ì†Œìš” ì‹œê°„: {total_time:.2f}s / í‰ê·  FPS: {fps:.2f}")

# ì˜ˆì¸¡ ê²°ê³¼ ë™ì˜ìƒ ê²½ë¡œ
result_dir = "runs/detect/predict"
output_files = glob.glob(os.path.join(result_dir, "*.mp4"))

# ì˜ˆì¸¡ ì˜ìƒì— ì˜¤ë¥¸ìª½ ìƒë‹¨ FPS í‘œì‹œ ì¶”ê°€
def add_fps_to_video(video_path, fps, output_path):
    cap = cv2.VideoCapture(video_path)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    input_fps = cap.get(cv2.CAP_PROP_FPS) or 30.0
    out = cv2.VideoWriter(output_path, fourcc, input_fps, (width, height))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        text = f'FPS: {fps:.2f}'
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 1
        color = (0, 255, 0)  # ì´ˆë¡ìƒ‰
        thickness = 2
        text_size, _ = cv2.getTextSize(text, font, font_scale, thickness)
        position = (frame.shape[1] - text_size[0] - 10, 30)  # ì˜¤ë¥¸ìª½ ìƒë‹¨
        cv2.putText(frame, text, position, font, font_scale, color, thickness)
        out.write(frame)
    cap.release()
    out.release()

# FPS ë°•ìŠ¤ ì²˜ë¦¬ëœ íŒŒì¼ ì €ì¥ ë° ë‹¤ìš´ë¡œë“œ
for i, video_file in enumerate(output_files):
    output_file = os.path.join(result_dir, f'fps_annotated_{i}.mp4')
    add_fps_to_video(video_file, fps, output_file)
    files.download(output_file)  # Colabì—ì„œ ë°”ë¡œ ë‹¤ìš´ë¡œë“œ

```
