```
import time
from ultralytics import YOLO
from google.colab import files
import glob
import os
import cv2

# Î™®Îç∏ ÌååÏùº ÏÑ†ÌÉù (Ïòà: 'yolov11n.pt', 'yolov12n.pt')
MODEL_PATH = 'yolo11n.pt'  # ÎòêÎäî 'yolov11n.pt'
model = YOLO(MODEL_PATH)
model.info()

# ÎèôÏòÅÏÉÅ ÌååÏùº ÏóÖÎ°úÎìú
uploaded = files.upload()
video_path = list(uploaded.keys())[0]

# FPS Ï∏°Ï†ïÏö© Ï∂îÎ°† Ïã§Ìñâ
start = time.time()
results = model.predict(video_path, save=True)
end = time.time()

# ÌîÑÎ†àÏûÑ Ïàò Î∞è ÏÜåÏöî ÏãúÍ∞Ñ
cap = cv2.VideoCapture(video_path)
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
cap.release()
total_time = end - start
fps = frame_count / total_time if total_time > 0 else 0
print(f"{MODEL_PATH} / ÌîÑÎ†àÏûÑ Ïàò: {frame_count} / Ï¥ù ÏÜåÏöî ÏãúÍ∞Ñ: {total_time:.2f}s / ÌèâÍ∑† FPS: {fps:.2f}")

# ÏòàÏ∏° Í≤∞Í≥º ÎèôÏòÅÏÉÅ Í≤ΩÎ°ú
result_dir = "runs/detect/predict"
output_files = glob.glob(os.path.join(result_dir, "*.mp4"))

# ÏòàÏ∏° ÏòÅÏÉÅÏóê Ïò§Î•∏Ï™Ω ÏÉÅÎã® FPS ÌëúÏãú Ï∂îÍ∞Ä
def add_fps_to_video(video_path, fps, output_path):
    cap = cv2.VideoCapture(video_path)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    input_fps = cap.get(cv2.CAP_PROP_FPS) or 30.0
    out = cv2.VideoWriter(output_path, fourcc, input_fps, (width, height))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        text = f'FPS: {fps:.2f}'
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 1
        color = (0, 255, 0)  # Ï¥àÎ°ùÏÉâ
        thickness = 2
        text_size, _ = cv2.getTextSize(text, font, font_scale, thickness)
        position = (frame.shape[1] - text_size[0] - 10, 30)  # Ïò§Î•∏Ï™Ω ÏÉÅÎã®
        cv2.putText(frame, text, position, font, font_scale, color, thickness)
        out.write(frame)
    cap.release()
    out.release()

# FPS Î∞ïÏä§ Ï≤òÎ¶¨Îêú ÌååÏùº Ï†ÄÏû• Î∞è Îã§Ïö¥Î°úÎìú
for i, video_file in enumerate(output_files):
    output_file = os.path.join(result_dir, f'fps_annotated_{i}.mp4')
    add_fps_to_video(video_file, fps, output_file)
    files.download(output_file)  # ColabÏóêÏÑú Î∞îÎ°ú Îã§Ïö¥Î°úÎìú

```
